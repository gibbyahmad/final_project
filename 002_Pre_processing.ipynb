```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
# Load data
data = pd.read_csv("./Data/train-flattened.csv")
#test_df = pd.read_csv("./Data/test-flattened.csv")
```

```python
data['date'] = data['date'].apply(lambda x: pd.to_datetime(str(x), format='%Y%m%d'))
```

```python
# Label encoder
object_col = [i for i in data.columns if data[i].dtype == 'object']
#object_col.remove(data['fullVisitorId'])
for col in object_col:
    print (col)
    train_unique = data[col].unique()
    train_dict = {k:v for v,k in enumerate(train_unique)}
    data[col] = data[col].apply(lambda x: train_dict.get(x))
```

```python
data.fillna(0,inplace=True)
data.tail()
```
